# Indian Premier League (IPL) Statistics - Exploratory Data Analysis

## 1. Introduction
##### What is the IPL?
The IPL is a professional men's Twenty20 cricket league founded in 2007 by the Board of Control for Cricket in India (BCCI). It has 10 competing teams based out of seven Indian cities and three Indian states.
The competition runs between March and May every year. It is the most-attended cricketing league in the world and in 2014 it was the sixth most attended sporting league in the world. It was the first league to be broadcast live on Youtube.
The IPL brand is thought to be valued upwards of US$6,000,000,000 (6 billion). It has an average of roughly 31,000,000 (31 million) impressions and continues to grow.

##### What will this exploratory analysis accomplish?
I will explore the importance of winning the toss and whether the choice to bat or bowl first matters, I will see if the venue provides any advantage to batting or bowling first and, I will see some fun facts out about top run scorers and wicket takers/.

##### Victory types at each venue
Ties are such a rare occurence, I shall remove them so the proportions are strictly runs vs wickets

```python
win_type = ("runs", "wickets")
no_ties = df_match[df_match["result"].isin(win_type)]
```

```python
# Prepare data for one bar that will show the victory type of "runs"
result_run = no_ties[no_ties["result"] == "runs"]
result_run = result_run.groupby(["result", "venue"])["id"].agg("count").reset_index()
result_run_sort = result_run.sort_values(["id"], ascending = False)
```

```python
# https://stackoverflow.com/questions/46623583/seaborn-countplot-order-categories-by-count

# Prepare to have the venues with the most matches played at the top of the chart
order = df_match["venue"].value_counts().index
```

```python
# https://www.python-graph-gallery.com/stacked-and-percent-stacked-barplot 

bar1 = sns.countplot(y = df_match["venue"], order = order, color = "r")
bar2 = sns.barplot(data = result_run_sort, y = "venue", x = "id", order = order, color = "b")

# Set up legend properties
top_bar = mpatches.Patch(color = "b", label = "Runs")
bot_bar = mpatches.Patch(color = "r", label = "Wickets")
plt.legend(handles = [top_bar, bot_bar])

plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8)

plt.xlabel("Matches played", fontsize = 12)
plt.ylabel("Venue", fontsize = 12)
plt.title('Proportion of victory types at each venue', fontweight = "bold", fontsize = 14)
```

All venues are roughly 50-50 in terms of victory type. Venues such as: Sawai Mansing stadium, M Chinnaswamy stadium, and Eden Gardens are minor skewed in favour of one victory type
